```python
numerical = ["Defense", "Attack", "Speed","Special Attack"]
numerical_axis = ["Defense", "Attack","Speed of pokemon", "Special Attack"]
```

```python
list(zip(numerical,numerical_axis))
```

```python
title_font ={"family":"arial","color":"darkred","weight":"bold","size":15}
axis_font ={"family":"arial","color":"darkgreen","weight":"bold","size":10}

for i,z in list(zip(numerical,numerical_axis)):
    plt.figure(figsize=(5,5),dpi=90)
    sms.distplot(df[i], hist_kws = dict(linewidth=1,edgecolor = 'r'))
    
    plt.title(i, fontdict = title_font)
    plt.xlabel(z, fontdict = axis_font)
    plt.ylabel("Density", fontdict = axis_font)
    
    plt.tight_layout()
    plt.show()
```

```python
title_font ={"family":"arial","color":"darkred","weight":"bold","size":15}
axis_font ={"family":"arial","color":"darkgreen","weight":"bold","size":10}

for i,z in list(zip(numerical,numerical_axis)):
    plt.figure(figsize=(5,5),dpi=90)
    sms.displot(df[i])
    
    plt.title(i, fontdict = title_font)
    plt.xlabel(z, fontdict = axis_font)
    plt.ylabel("Density", fontdict = axis_font)
    
    plt.tight_layout()
    plt.show()
```

```python
title_font ={"family":"arial","color":"darkred","weight":"bold","size":15}
axis_font ={"family":"arial","color":"darkgreen","weight":"bold","size":10}

for i,z in list(zip(numerical,numerical_axis)):
    x,y = plt.subplots(figsize=(5,5))
    sms.distplot(df[i], hist = False,ax = y)
    y.axvline(df[i].mean(),color="r",ls="--");
    
    plt.title(i, fontdict = title_font)
    plt.xlabel(z, fontdict = axis_font)
    plt.ylabel("Density", fontdict = axis_font)
    
    plt.tight_layout()
    plt.show()
```
